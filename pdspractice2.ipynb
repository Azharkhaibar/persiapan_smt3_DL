```python
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns

# Data dasar
faculties = ["Teknik", "Ekonomi", "Kedokteran", "Hukum", "Sosial", "MIPA", "Sastra"]
programs = ["S1", "S2", "S3"]
campus_locations = ["Jakarta", "Bandung", "Surabaya", "Tangerang", "Yogyakarta", "Medan"]
graduation_gpas = [round(random.uniform(2.5, 4.0), 2) for _ in range(150)]
graduate_salaries = list(range(3000000, 20000000, 1000000))
years_of_graduation = list(range(2015, 2024))
genders = ["Male", "Female"]
employment_statuses = ["Employed", "Unemployed", "Freelancer", "Entrepreneur"]
publications = list(range(0, 20))

# Membuat data lulusan
graduates = {
    "Fakultas": [random.choice(faculties) for _ in range(150)],
    "Program_Studi": [random.choice(programs) for _ in range(150)],
    "Lokasi_Kampus": [random.choice(campus_locations) for _ in range(150)],
    "Tahun_Lulus": [random.choice(years_of_graduation) for _ in range(150)],
    "Jenis_Kelamin": [random.choice(genders) for _ in range(150)],
    "Status_Pekerjaan": [random.choice(employment_statuses) for _ in range(150)],
    "IPK_Lulusan": graduation_gpas,
    "Gaji_Lulusan": [random.choice(graduate_salaries) for _ in range(150)],
    "Jumlah_Publikasi": [random.choice(publications) for _ in range(150)]
}

# Membuat DataFrame
univ_df = pd.DataFrame(graduates)
univ_df.columns = [
    "FAKULTAS", "PROGRAM_STUDI", "LOKASI_KAMPUS", "TAHUN_LULUS", "JENIS_KELAMIN",
    "STATUS_PEKERJAAN", "IPK_LULUSAN", "GAJI_LULUSAN", "JUMLAH_PUBLIKASI"
]

# Menampilkan DataFrame
print(univ_df.head())

# 1. Menampilkan Statistik Deskriptif untuk Data Utama
print("\nStatistik Deskriptif untuk IPK Lulusan:")
print(univ_df["IPK_LULUSAN"].describe())

print("\nStatistik Deskriptif untuk Gaji Lulusan:")
print(univ_df["GAJI_LULUSAN"].describe())

print("\nStatistik Deskriptif untuk Jumlah Publikasi:")
print(univ_df["JUMLAH_PUBLIKASI"].describe())

# 2. Visualisasi Distribusi IPK dan Gaji Lulusan
plt.figure(figsize=(12, 6))
sns.histplot(univ_df["IPK_LULUSAN"], bins=10, kde=True, color='blue')
plt.title('Distribusi IPK Lulusan')
plt.xlabel('IPK')
plt.ylabel('Jumlah Lulusan')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(univ_df["GAJI_LULUSAN"], bins=10, kde=True, color='green')
plt.title('Distribusi Gaji Lulusan')
plt.xlabel('Gaji')
plt.ylabel('Jumlah Lulusan')
plt.show()

# 3. Analisis Gaji Berdasarkan Fakultas
gaji_fakultas = univ_df.groupby("FAKULTAS")["GAJI_LULUSAN"].agg(['mean', 'median', 'sum'])
print("\nGaji Berdasarkan Fakultas:")
print(gaji_fakultas)

# Visualisasi Gaji Berdasarkan Fakultas
plt.figure(figsize=(10, 6))
sns.barplot(x=gaji_fakultas.index, y=gaji_fakultas['mean'], palette='viridis')
plt.title('Rata-rata Gaji Lulusan Berdasarkan Fakultas')
plt.xlabel('Fakultas')
plt.ylabel('Rata-rata Gaji')
plt.show()

# 4. Korelasi Antara IPK, Gaji, dan Jumlah Publikasi
correlation_matrix = univ_df[["IPK_LULUSAN", "GAJI_LULUSAN", "JUMLAH_PUBLIKASI"]].corr()
print("\nKorelasi Antara IPK, Gaji, dan Jumlah Publikasi:")
print(correlation_matrix)

# Visualisasi Matriks Korelasi
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriks Korelasi IPK, Gaji, dan Jumlah Publikasi')
plt.show()

# 5. Analisis Status Pekerjaan Berdasarkan Jenis Kelamin
status_pekerjaan_gender = univ_df.groupby(["JENIS_KELAMIN", "STATUS_PEKERJAAN"]).size().unstack()
print("\nStatus Pekerjaan Berdasarkan Jenis Kelamin:")
print(status_pekerjaan_gender)

# Visualisasi Status Pekerjaan Berdasarkan Jenis Kelamin
status_pekerjaan_gender.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set3')
plt.title('Distribusi Status Pekerjaan Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah Lulusan')
plt.show()

# 6. Filter Data Berdasarkan Kriteria
# Lulusan dengan Gaji di Atas Rata-rata dan IPK di Atas 3.0
average_salary = univ_df["GAJI_LULUSAN"].mean()
high_salary_high_gpa_graduates = univ_df[(univ_df["GAJI_LULUSAN"] > average_salary) & (univ_df["IPK_LULUSAN"] > 3.0)]
print("\nLulusan dengan Gaji di Atas Rata-rata dan IPK di Atas 3.0:")
print(high_salary_high_gpa_graduates)

# Lulusan dengan Jumlah Publikasi di Atas 5
high_publication_graduates = univ_df[univ_df["JUMLAH_PUBLIKASI"] > 5]
print("\nLulusan dengan Jumlah Publikasi di Atas 5:")
print(high_publication_graduates)
```
